# Creating a new dataframe 
i created a new dataframe for the few data we will be needing for the logistic regression model

```python
new_df = df[['Accident_Severity','Number_of_Vehicles','Number_of_Casualties',]]
```

```python
new_df
```

```python
#from sklearn.preprocessing import MinMaxScaler
#scaler = MinMaxScaler()

features_df = new_df.drop(columns=['Accident_Severity'])
target = new_df['Accident_Severity']
x_train, x_test, y_train, y_test = train_test_split(features_df,target,test_size=0.3, random_state=2)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn import metrics
metrics.accuracy_score(y_pred,y_test)
```

```python
confusion_matrix = metrics.plot_confusion_matrix(model, x_test, y_test, cmap='GnBu')
```
